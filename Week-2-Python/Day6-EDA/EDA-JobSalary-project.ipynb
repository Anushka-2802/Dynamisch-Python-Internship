```python
#univariate analysis 
#salary distribution
plt.hist(df["salary"],bins=30)
plt.title("salary distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
# Experience Distribution
plt.hist(df["experience_years"], bins=20)
plt.title("Experience Distribution")
plt.xlabel("Experience")
plt.ylabel("Employees")
plt.show()
```

```python
# education level
df["education_level"].value_counts().plot(kind="bar")
plt.title("Education Level")
plt.show()
```

```python
 #bivariate analysis
#average salary by industry
df.groupby("industry")["salary"].mean().plot(kind="bar")
plt.title("Average salary by industry")
plt.show()
```

```python
#top 10 highest paying job title
top=df.groupby("job_title")["salary"].mean().sort_values(ascending=False).head(10)
top.plot(kind="bar")
plt.title("Top 10 highest paying job titles")
plt.show()
```

```python
#scatter plot
#salary vs experience
plt.scatter(df["experience_years"],df["salary"])
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#highest salary
df.nlargest(5,"salary")
```

```python
df.nsmallest(5,"salary")
```

```python
#outlier detection
plt.boxplot(df["salary"])
plt.title("Salary boxplot")
plt.show()
```
